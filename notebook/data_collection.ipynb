```python
import requests
import json

def get_api_session_key(api_url, access_code, secret_code):
    """Authenticate and get the session key."""
    url = f"{api_url}/login?accessCode={access_code}&secretCode={secret_code}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json().get('sessionKey')
    else:
        # Print status code and error message from API if available
        error_info = response.text
        raise Exception(f"Failed to authenticate with API: Status {response.status_code}, Error: {error_info}")

def prettify_json(json_data):
    """Convert a dictionary into a pretty-printed JSON string."""
    return json.dumps(json_data, indent=4)

def get_updated_artists(api_url, session_key, pagination_token=None):
    """Fetch the list of updated artists and return a prettified JSON string."""
    url = f"{api_url}/UpdatedArtists"
    params = {
        'authSessionKey': session_key,
        'paginationToken': pagination_token
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        json_data = response.json()
        return prettify_json(json_data)
    else:
        error_info = response.text  # Providing more detail in case of an error
        raise Exception(f"Failed to fetch updated artists: Status {response.status_code}, Error: {error_info}")


# API base URL
api_url = "https://www.wikiart.org/en/Api/2"

# Your API access credentials
access_code = "d19a1905694a438f"
secret_code = "93ce65ee0976e0a2"

# Authenticate and get a session key
try:
    session_key = get_api_session_key(api_url, access_code, secret_code)
    # Get the first page of updated artists
    artists_data = get_updated_artists(api_url, session_key)
    print(artists_data)
except Exception as e:
    print(e)

# Note: Replace YOUR_ACCESS_CODE and YOUR_SECRET_CODE with your actual credentials.
```
