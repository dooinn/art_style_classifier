```python
import os

def count_images_in_directories(parent_directory):
    # Define possible image extensions
    image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp'}
    
    # Dictionary to hold the count of images in each directory
    directory_image_count = {}
    
    # Walk through the directory structure
    for root, dirs, files in os.walk(parent_directory):
        # Count images only in subdirectories of the parent directory
        if root != parent_directory:
            count = 0
            for file in files:
                # Check if the file is an image based on its extension
                if os.path.splitext(file)[1].lower() in image_extensions:
                    count += 1
            # Add the count to the dictionary
            directory_name = os.path.basename(root)
            directory_image_count[directory_name] = count
    
    return directory_image_count

# Specify the path to the parent directory
parent_directory = '../data/images'
# Get the count of images in each directory
image_counts = count_images_in_directories(parent_directory)
# Print the results
for directory, count in image_counts.items():
    print(f"{directory}: {count} images")

```

```python
import os
import shutil
import random

def select_random_images(source_directory, destination_directory, num_images=100):
    # Define possible image extensions
    image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp'}
    
    # Ensure the destination directory exists
    os.makedirs(destination_directory, exist_ok=True)
    
    # Walk through the directory structure
    for root, dirs, files in os.walk(source_directory):
        # Filter and list full image paths
        images = [os.path.join(root, file) for file in files if os.path.splitext(file)[1].lower() in image_extensions]
        
        # If the number of images is more than required, randomly select the images
        if len(images) > num_images:
            selected_images = random.sample(images, num_images)
        else:
            selected_images = images
        
        # Create a destination subdirectory
        genre_name = os.path.basename(root)
        genre_dir = os.path.join(destination_directory, genre_name)
        os.makedirs(genre_dir, exist_ok=True)
        
        # Copy selected images to the new directory
        for image in selected_images:
            shutil.copy(image, genre_dir)
        print(f"Copied {len(selected_images)} images to {genre_dir}")

# Specify the source and destination directories
source_directory = '../data/images'
destination_directory = '../data/selected_images'

# Run the function
select_random_images(source_directory, destination_directory)
```

```python
import os
import shutil
import random

def split_dataset(source_directory, base_output_directory, train_ratio=0.8, validation_ratio=0.1, test_ratio=0.1):
    # Define possible image extensions
    image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp'}
    
    # Walk through the directory structure
    for root, dirs, files in os.walk(source_directory):
        # Filter and list full image paths
        images = [os.path.join(root, file) for file in files if os.path.splitext(file)[1].lower() in image_extensions]
        random.shuffle(images)  # Shuffle images to ensure random distribution
        
        # Compute split indices
        total_images = len(images)
        train_end = int(total_images * train_ratio)
        validation_end = train_end + int(total_images * validation_ratio)
        
        # Split images based on calculated indices
        train_images = images[:train_end]
        validation_images = images[train_end:validation_end]
        test_images = images[validation_end:]
        
        # Subdirectory paths
        genre_name = os.path.basename(root)
        train_dir = os.path.join(base_output_directory, 'train', genre_name)
        validation_dir = os.path.join(base_output_directory, 'validation', genre_name)
        test_dir = os.path.join(base_output_directory, 'test', genre_name)
        
        # Create directories
        os.makedirs(train_dir, exist_ok=True)
        os.makedirs(validation_dir, exist_ok=True)
        os.makedirs(test_dir, exist_ok=True)
        
        # Function to copy images to respective directories
        def copy_images(images, directory):
            for image in images:
                shutil.copy(image, directory)
        
        # Copy images to respective directories
        copy_images(train_images, train_dir)
        copy_images(validation_images, validation_dir)
        copy_images(test_images, test_dir)
        
        print(f"Processed {genre_name}:")
        print(f"  Training images: {len(train_images)}")
        print(f"  Validation images: {len(validation_images)}")
        print(f"  Test images: {len(test_images)}")

# Specify the source directory and the base directory for output datasets
source_directory = '../data/images/'
base_output_directory = 'dataset'

# Run the function
split_dataset(source_directory, base_output_directory)
```
